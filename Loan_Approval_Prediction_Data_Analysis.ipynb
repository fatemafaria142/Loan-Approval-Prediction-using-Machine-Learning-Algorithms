```python
label_counts = df["Residence_State"].value_counts()

# Define custom colors for the bars
custom_colors = ['#e34861' , '#2595b0','#84d0ff', '#846f6f', '#f7f733', '#ffbb84', '#2889ff', '#f989ff', '#b5d1ff', '#a789ff',
                 '#66d6ff', '#ff8e66', '#ff00ee', '#ff1c2e', '#0058fc', '#ff8077','#99269b','#87ba12', '#ea7115','#ffbcbc',
                 '#5775c1', '#6afcb8','#368141','#067072','#a5a5a5','#05223a','#7a8c00','#e34861']

# Define custom font dictionary for title and labels
font = {'family': 'Serif', 'weight': 'bold', 'size': 24}
font2 = {'family': 'Serif', 'weight': 'bold', 'size': 18}

plt.figure(figsize=(38, 14))

# Create bar plot with grid
bars = plt.bar(label_counts.index, label_counts.values, color=custom_colors)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set title and axis labels using custom fontdict
plt.title('Residence State Distribution', fontdict=font,pad=10)
plt.xlabel('Residence State', fontdict=font2)
plt.ylabel('Count', fontdict=font)

# Set custom font for ticks on both x and y axes with rotation
plt.xticks(label_counts.index, label_counts.index, fontdict=font, rotation=90)
plt.yticks(fontname='Serif', fontsize=18)

# Adding annotations (count values) on top of each bar
for bar, count in zip(bars, label_counts.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(count),
             ha='center', va='bottom', fontdict=font2)

# Show the plot
plt.show()
```

#### **Distribution of Occupations in the Dataset**

```python
# Set custom font dictionary for title and labels
font = {'family': 'Serif', 'weight': 'bold', 'size': 24}
font2 = {'family': 'Serif', 'weight': 'bold', 'size': 18}

# Set figure size
plt.rcParams["figure.figsize"] = (24, 12)

# Create a bar plot for Occupation distribution
ax = sns.countplot(data=df, x="Occupation", color='#e34861')  # You can adjust the color as needed

# Annotate counts on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 35), textcoords='offset points', fontsize=16, rotation=90)

# Set title and axis labels using custom fontdict
plt.title('Distribution of Occupations in the Dataset', fontdict=font, pad=12)
plt.xlabel('Occupation', fontdict=font2)
plt.ylabel('Count', fontdict=font2)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90, ha='right', fontsize=16)

# Display the plot
plt.show()

```
